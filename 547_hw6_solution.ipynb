# 1. controllability of interconnections

Consider the following schematics of an RC circuit and spring-mass-damper system:

![](https://github.com/sburden/547-21wi/raw/master/fig_11.1_RC-circuit_suspension.PNG)

Letting $x\in\mathbb{R}^2$ denote capacitor voltages and $u\in\mathbb{R}$ denote voltage source in Figure (a) yields dynamics

$$ \dot{x} = \left[\begin{array}{cc} -\frac{1}{R_1 C_1} & 0 \\ 0 & -\frac{1}{R_2 C_2} \end{array}\right] x + \left[\begin{array}{c} \frac{1}{R_1 C_1} \\ \frac{1}{R_2 C_2} \end{array}\right] u. $$

Assume that the time constants are equal, $\frac{1}{R_1 C_1} = \frac{1}{R_2 C_2} = \omega$.

(a) *Determine the reachable subspace of circuit voltages.*

***Solution:*** given initial state $x(0)$ and input $u:[0,t]\rightarrow\mathbb{R}$ and with $\omega = 1/R_1 C_1 = 1/R_2 C_2$ we have

$$ x(t) = e^{-\omega t} x(0) + \omega \int_0^t e^{-\omega(t-\tau)} u(\tau) d\tau \left[\begin{array}{c} 1 \\ 1 \end{array}\right], $$

so the subspace of states reachable from the origin $x(0) = 0$ is spanned by the vector of ones,

$$ \mathcal{R}[0,t] = \left\{ \alpha \left[\begin{array}{c} 1 \\ 1 \end{array}\right] : \alpha\in\mathbb{R} \right\}. $$



Letting $x\in\mathbb{R}^4$ denote mass positions and velocities and $u\in\mathbb{R}$ denote load on the suspension system in Figure (b) yields dynamics

$$ \dot{x} = \left[\begin{array}{cccc} -\frac{b_1}{m_1} & -\frac{k_1}{m_1} & 0 & 0 \\ 1 & 0 & 0 & 0 \\ 0 & 0 & -\frac{b_2}{m_2} & -\frac{k_2}{m_2} \\ 0 & 0 & 1 & 0 \end{array}\right] x + \left[\begin{array}{c} \frac{1}{2 m_1} \\ 0 \\ \frac{1}{2 m_2} \\ 0 \end{array}\right] u. $$

Assume that the suspension system is symmetric, $m_1 = m_2$, $b_1 = b_2$, $k_1 = k_2$.

(b) *Determine the reachable subspace of mass positions and velocities.*

***Solution:*** we will derive the reachable subspace for the general case of parallel interconnection of identical subsystems (of which this mechanical system and the circuit in (a.) are special cases).

To that end, consider the LTI-DE with state $x = (x_1, x_2)\in\mathbb{R}^{2n}$,

$$ \dot{x} = \left[\begin{array}{cc} A_1 & 0 \\ 0 & A_2 \end{array}\right] x + \left[\begin{array}{c} B_1 \\ B_2 \end{array}\right] u, $$

whose trajectories are given by

$$ x(t) = \left[\begin{array}{c} e^{A_1 t} x_1(0) \\ e^{A_2 t} x_2(0) \end{array}\right] + \int_0^t \left[\begin{array}{c}  e^{A_1 (t-\tau)} B_1 \\ e^{A_2 (t-\tau)} B_2 \end{array}\right] u(\tau) d\tau. $$

When $A_1 = A_2 = A$ and $B_1 = B_2 = B$ and $x(0) = 0$ we have

$$ x(t) = \left[\begin{array}{c} I \\ I \end{array}\right]\int_0^t e^{A (t-\tau)} B u(\tau) d\tau, $$

so the reachable subspace is clearly

$$ \mathcal{R}[0,t] = \left\{ \left[\begin{array}{c} \xi \\ \xi \end{array}\right] : \xi\in\mathbb{R}^n \right\}. $$

In the context of the mechanical system, the reachable subspace consists of states where the position and velocity of the left- and right-half of the suspensions system are equal.

Now consider the following block diagrams showing interconnections between subsystems:

![](https://github.com/sburden/547-21wi/raw/master/fig_11.2_block-interconnections.PNG)

(c) *With $x = (x_1,x_2)$, determine matrices $\bar{A}$ and $\bar{B}$ such that $\dot{\bar{x}} = \bar{A} \bar{x} + \bar{B} u$ for the parallel interconnection.*

***Solution:***

$$ \dot{\bar{x}} = \left[\begin{array}{cc} A_1 & 0 \\ 0 & A_2 \end{array}\right] \bar{x} + \left[\begin{array}{c} B_1 \\ B_2 \end{array}\right] \bar{u} = \bar{A} \bar{x} + \bar{B} \bar{u}. $$



(d) *With $\bar{x} = (x_1,x_2)$, determine matrices $\bar{A}$ and $\bar{B}$ such that $\dot{\bar{x}} = \bar{A} \bar{x} + \bar{B} u$ for the series interconnection; assume $z = x_1$ is the input to the second block (i.e. the "$u$" in the second block equals $x_1$).*

***Solution:***

$$ \dot{\bar{x}} = \left[\begin{array}{cc} A_1 & 0 \\ B_2 & A_2 \end{array}\right] \bar{x} + \left[\begin{array}{c} B_1 \\ 0 \end{array}\right] \bar{u} = \bar{A} \bar{x} + \bar{B} \bar{u}. $$



Now assume $A_1 = A_2 = A$, $B_1 = B_2 = B$.

(e) *If $(A,B)$ is controllable, is the parallel interconnection in (c) controllable?*

***Solution:*** no -- the suspension system from (b) provides a counterexample:  since this system consists of a parallel interconnection of identical subsystems, and since the set of reachable states was a strict subspace (i.e. not equal to the entire state space), parallel interconnection of identical subsystems yields an uncontrollable interconnected system.



***Bonus (f):*** *If $(A,B)$ is controllable, is the series interconnection in (d) controllable?*

***Note:*** *I changed this to a "bonus" problem because the only solution I could come up with requires techniques we haven't covered yet.*

***Solution:*** we are given that (A,B) is controllable, and we are treating the output from the first subsystem as the full state, i.e. $C = I$, $D = 0$.  Thus, both subsystems are completely controllable *and* observable, i.e. they are *minimal* representations of their transfer matrices.  Noting that no pole/zero cancellations occur in the individual subsystems (minimial representations do not have internal pole/zero cancellations) or in their interconnection (multiplying identical transfer matrices does not cause pole/zero cancellations), we conclude the state-space realization in (d) is minimal, so the interconnected system is fully controllable (and observable).

***Note:*** if the subsystems are not identical, it is possible that a zero of the first cancels a pole of the second, leading to a loss of controllability or observability of the interconnected system.



# 2. minimum-"energy" control

Suppose $x_1$ is reachable from the origin on time interval $[t_0,t_1]$ in (CT-LTV-DE) $\dot{x} = A(t) x + B(t) u$.

Let $x_1 = W(t_0,t_1) \eta_1$ and $u(t) = B(t)^\top \Phi(t_1,t)^\top \eta_1$ for $t\in[t_0,t_1]$; recall that that $u$ steers $x(t_0) = 0$ to $x(t_1) = x_1$.

Suppose $\bar{u}$ is another input that steers $x(t_0) = 0$ to $x(t_1) = x_1$ and let $v = u-\bar{u}$.

(a) *Compute $\int_{t_0}^{t_1} \Phi(t_1,\tau) B(\tau) v(\tau) d\tau$.*

***Solution:*** 

$$ \int_{t_0}^{t_1} \Phi(t_1,\tau) B(\tau) v(\tau) d\tau = \int_{t_0}^{t_1} \Phi(t_1,\tau) B(\tau) u(\tau) d\tau - \int_{t_0}^{t_1} \Phi(t_1,\tau) B(\tau) \bar{u}(\tau) d\tau = x_1 - x_1 = 0. $$



(b) *Show that $\int_{t_0}^{t_1} \| u(\tau) \|^2 d\tau = \eta_1^\top W(t_0,t_1) \eta_1$ using the fact that $\| w \|^2 = w^\top w$.*

***Solution:*** using the given facts:

$$ \int_{t_0}^{t_1} \| u(\tau) \|^2 d\tau = \int_{t_0}^{t_1} u(\tau)^\top u(\tau) d\tau = \int_{t_0}^{t_1} \eta_1^\top \Phi(t_1,\tau) B(\tau) B(\tau)^\top \Phi(t_1,\tau)^\top \eta_1 d\tau = \eta_1^\top W(t_0,t_1) \eta_1. $$


(c) *Show that $\int_{t_0}^{t_1} \| \bar{u}(\tau) \|^2 d\tau = \int_{t_0}^{t_1} \| u(\tau) \|^2 d\tau + \int_{t_0}^{t_1} \| v(\tau) \|^2 d\tau$.*

***Solution:*** applying the facts from (a.) and (b.):

$$ \int_{t_0}^{t_1} \| \bar{u}(\tau) \|^2 d\tau = \int_{t_0}^{t_1} \| B(t)^\top \Phi(t_1,t)^\top \eta_1 + v(\tau) \|^2 d\tau d\tau $$

$$ = \int_{t_0}^{t_1} \| u(\tau) \|^2 + \int_{t_0}^{t_1} \Phi(t_1,\tau) B(\tau) v(\tau) d\tau + \int_{t_0}^{t_1} \| v(\tau) \|^2 $$

$$ = \int_{t_0}^{t_1} \| u(\tau) \|^2 d\tau + \int_{t_0}^{t_1} \| v(\tau) \|^2 d\tau. $$

***Conclusion:*** $u$ is the minimum-"energy" (really:  minimum-norm) control input that steers $x(t_0) = 0$ to $x(t_1) = x_1$.



# 3. controllable decomposition

Consider the CT-LTI system specified by

$$ A = \left[\begin{array}{cc} -1 & 0 \\ 0 & -1 \end{array}\right],\ B = \left[\begin{array}{c} -1 \\ 1 \end{array}\right],\ C = \left[\begin{array}{cc} 1 & 0 \\ 0 & 1 \end{array}\right],\ D = \left[\begin{array}{c} 2 \\ 1 \end{array}\right]. $$

(a) Is CT-LTI controllable?

***Solution:*** we'll let `SymPy` do the heavy lifting :)



```
import sympy as sym

A = sym.Matrix([[-1,0],[0,-1]])
B = sym.Matrix([[-1],[1]])
C = sym.Matrix([[1,0],[0,1]])
D = sym.Matrix([[2],[1]])

print(A)
print(B)
```

```
Co = sym.Matrix.hstack(B, A @ B)
print(Co)
```

```
Co.rank()
```

Since the controllability matrix is not full-rank, we conclude the system is not controllable.

(b) Apply the procedure derived in problem 2 on HW 5 to perform a ***controllable decomposition*** on the system.

***Solution:***  the first column of the controllability matrix spans its range space; we choose a second vector that is obviously linearly independent to complete the basis:

$$ T = \left[\begin{array}{cc} -1 & 0 \\ 1 & 1 \end{array}\right] $$

We'll let `SymPy` do the matrix multiplication :)



```
T = sym.Matrix([[-1,0],[1,1]])
T
```

$T$ turns out to be [involutary](https://en.wikipedia.org/wiki/Involutory_matrix), i.e. $T$ is its own inverse:

```
TI = sym.Matrix.inv(T)
TI
```

```
TI @ A @ T
```

```
TI @ B
```

(c) Use the result from (b) to create a controllable state-space model that has the same transfer matrix as CT-LTI.

***Solution:*** pulling out the controllable subsystem (i.e. top-left entry in $T^{-1} A T$, top entry in $T^{-1} B$, left column in $C T$) yields

$$ A_c = -1, B_c = 1, C_c = \left[\begin{array}{cc} 1 \\ -1 \end{array}\right], D_c = D. $$

```
Ac = (TI @ A @ T)[:1,:1]
Ac
```

```
Bc = sym.Matrix((TI @ B)[:1])
Bc
```

```
Cc = (C @ T)[:,:1]
Cc
```

```
Dc = D
```

We verify these matrices yield the same transfer matrix:

```
s = sym.symbols(r's')
```

```
G = C @ sym.Matrix.inv(s*sym.eye(2) - A) @ B + D
G
```

```
Gc = Cc @ sym.Matrix.inv(s*sym.eye(1) - Ac) @ Bc + Dc
Gc
```

```
G == Gc
```
