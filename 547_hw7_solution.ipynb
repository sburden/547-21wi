# 1. observability via duality

In this problem, you will derive ***observability*** tests for the ${\text{LTI-DE}}$

$$ \dot{x}/x^+ = A x + B u,\ x\in\mathbb{R}^n,\ u\in\mathbb{R}^k, $$

$$ y = C x + D u,\ y\in\mathbb{R}^m, $$

from ***controllability*** tests for the ***dual*** $\overline{\text{LTI-DE}}$

$$ \dot{\bar{x}}/\bar{x}^+ = A^\top \bar{x} + C^\top \bar{u},\ \bar{x}\in\mathbb{R}^n,\ \bar{u}\in\mathbb{R}^m, $$

$$ \bar{y} = B^\top \bar{x} + D^\top \bar{u},\ \bar{y}\in\mathbb{R}^k. $$

Define the ***observability matrix*** for ${\text{LTI-DE}}$ to be the transpose of the controllability matrix $\mathcal{\bar{C}}$ for $\overline{\text{LTI-DE}}$, $\mathcal{O} = \mathcal{\bar{C}}^\top$.

(a.) Apply the controllability matrix rank test for controllability of $\overline{\text{LTI-DE}}$ to obtain the observability matrix rank test for observability of ${\text{LTI-DE}}$.

***Solution:*** ${\text{LTI-DE}}$ is observable if the ***observability matrix*** 

$$ \mathcal{O} = \bar{\mathcal{C}}^\top = \left[\begin{array}{c} C \\ C A \\ C A^2 \\ \vdots \\ C A^{n-1} \end{array}\right] $$

has rank $n$.



(b.) Apply the eigenvector test for controllability of $\overline{\text{LTI-DE}}$ to obtain the eigenvector test for observability of ${\text{LTI-DE}}$.

***Solution:*** ${\text{LTI-DE}}$ is observable if no eigenvector of $A$ is in the nullspace of $C$.



(c.) Apply the PBH test for controllability of $\overline{\text{LTI-DE}}$ to obtain the PBH test for observability of ${\text{LTI-DE}}$.

***Solution:*** ${\text{LTI-DE}}$ is observable if 

$$ \forall\lambda\in\mathbb{C} : \operatorname{rank}\left[\begin{array}{c} A - \lambda I \\ C \end{array}\right] = n. $$



(d.) Apply the Lyapunov test for controllability of $\overline{\text{LTI-DE}}$ to obtain the Lyapunov test for observability of ${\text{LTI-DE}}$.

***Solution:*** ${\text{LTI-DE}}$ is observable if there is a unique positive-definition solution $M$ to the Lyapunov equation

$$ (\text{CT})\ A^\top M + M A - - C^\top C, $$
$$ (\text{DT})\ A^\top M A - M = -C^\top C. $$

# 2. observable *and* controllable decomposition

Consider the LTI-DE 

$$ \dot{x}/x^+ = A x + B u,\ x\in\mathbb{R}^n,\ u\in\mathbb{R}^k, $$

$$ y = C x + D u,\ y\in\mathbb{R}^m. $$

(a.) Use duality to show that $(A,C)$ is observable if and only if $(T^{-1} A T, C T)$ is observable.

***Solution:*** 
duality ensures that $(A,C)$ is observable $\iff$ (if and only if) $(A^\top,C^\top)$ is controllable.  
Problem 2a from Homework 5 ensures $(A^\top,C^\top)$ is controllable $\iff$ $(T^\top A^\top (T^{-1})^\top, T^\top C^\top)$ is controllable.  
Duality ensures that $(T^\top A^\top (T^{-1})^\top, T^\top C^\top)$ is controllable $\iff$ $(T^{-1} A T, C T)$ is observable.  
Chaining the three previous "if-and-only-if" statements yields the desired result.


(b.) Use duality to derive the ***observable decomposition*** of the LTI-DE -- that is, show that there exists an invertible matrix $T$ and matrices $A_o, A_{o\bar{o}}, A_\bar{o}, B_o, B_\bar{o}, C_o$ such that

$$ \tilde{A} = T^{-1} A T = \left[\begin{array}{cc} A_o & 0 \\ A_{o\bar{o}} & A_\bar{o} \end{array}\right],\ 
\tilde{B} = T^{-1} B = \left[\begin{array}{cc} B_o \\ B_\bar{o} \end{array}\right],\ 
\tilde{C} = C T = \left[\begin{array}{cc} C_o & 0 \end{array}\right], $$

where $(A_o,C_o)$ is observable; be sure to specify the dimensions of $A_o, B_o, C_o$.

***Solution:*** 
Problem 2d from Homework 5 shows that, if 

$\bar{V}\in\mathbb{R}^{n\times\eta}$ is such that $\mathcal{R}(\bar{V}) = \mathcal{R}(\bar{\mathcal{C}})$ where $\bar{\mathcal{C}}$ is the controllability matrix corresponding to the pair $(\bar{A},\bar{B})$

and
$\bar{U}\in\mathbb{R}^{n\times(n-\eta)}$ is such that

$$ \bar{T} = \left[\begin{array}{cc} \bar{V} & \bar{U} \end{array}\right] $$

is nonsingular, then with

$$ \bar{T}^{-1} \bar{A} \bar{U} = \left[\begin{array}{c} \bar{A}_{c\bar{c}} \\ \bar{A}_\bar{c} \end{array} \right],\ \bar{A}_{c\bar{c}}\in\mathbb{R}^{\eta\times(n-\eta)},\ \bar{A}_\bar{c}\in\mathbb{R}^{(n-\eta)\times(n-\eta)}, $$

we have that $(\bar{A}_c,\bar{B}_c)$ is controllable and

$$ \bar{T}^{-1} \bar{A} \bar{T} = \left[\begin{array}{cc} \bar{A}_c & \bar{A}_{c\bar{c}} \\ 0 & \bar{A}_\bar{c} \end{array} \right],\ \bar{T}^{-1} \bar{B} = \left[\begin{array}{c} \bar{B}_c \\ 0 \end{array} \right],\ \bar{C}\bar{T} = \left[\begin{array}{c} \bar{C}_c \\ \bar{C}_\bar{c} \end{array} \right]. $$

Applying duality to the controllable decomposition of the dual system $(\bar{A},\bar{B},\bar{C}) = (A^\top,C^\top,B^\top)$ yields the desired result via invertible matrix $T = (\bar{T}^\top)^{-1}$.



(c.) Show that the ***non-observable subspace*** $\mathcal{N}$ of the ***observable decomposition*** from (c.) is given by

$$ \mathcal{N} = \mathcal{R}\left(\left[\begin{array}{c} 0 \\ I \end{array}\right]\right); $$

be sure to specify the dimensions of the zero matrix and identity matrix in this equation.

***Solution:*** the observability matrix of the ***observable decomposition*** is

$$ \tilde{\mathcal{O}} = \left[\begin{array}{c} \tilde{C} \\ \tilde{C} \tilde{A} \\ \tilde{C} \tilde{A}^2 \\ \vdots \\ \tilde{C} \tilde{A}^{n-1} \end{array}\right] = \left[\begin{array}{cc} C_o & 0_{m\times (n-\eta)} \\ C_o A_o & 0_{m\times (n-\eta)} \\ C_o A_o^2 & 0_{m\times (n-\eta)} \\ \vdots & \vdots \\ C_o A_o^{n-1} & 0_{m\times (n-\eta)} \end{array}\right] =  \left[\begin{array}{c} \mathcal{O}_o & 0_{mn\times(n-\eta)} \end{array}\right]. $$

Since $(A_o,C_o)$ is observable by construction, their observability matrix $\mathcal{O}_o$ has rank $n$, so the first $\eta$ columns of $\tilde{\mathcal{O}}$ are full-rank (have rank $\eta$); the last $n-\eta$ columns clearly have zero rank.  We conclude that the ***non-observable subspace*** of the ***observable decomposition*** is given by

$$ \mathcal{N} = \mathcal{N}(\tilde{\mathcal{O}}) = \mathcal{R}\left(\left[\begin{array}{c} 0_{\eta\times n} \\ I_{(n-\eta)\times n} \end{array}\right]\right). $$

(e.) Show that

$$ H^{-1} A H = \left[\begin{array}{cccc} A_{co} & 0 & A_{\times o} & 0 \\ A_{c\times} & A_{c\bar{o}} & A_{\times\times} & A_{\times\bar{o}} \\ 0 & 0 & A_{\bar{c}o} & 0 \\ 0 & 0 & A_{\bar{c}\times} & A_{\bar{c}\bar{o}} \end{array}\right],\ 
H^{-1} B = \left[\begin{array}{c} B_{co} \\ B_{c\bar{o}} \\ 0 \\ 0 \end{array}\right],\ 
C H = \left[\begin{array}{cccc} C_{co} & 0 & C_{\bar{c}o} & 0 \end{array}\right]. $$

***Solution:*** we derive this decomposition using similar techniques as the controllable (or observable) decomposition, that is, by noting that $\mathcal{C}$, $\mathcal{N}$, and $\mathcal{C}\cap\mathcal{N}$ are all $A$-invariant subspaces.  The key is noting that the key challenge in this problem is determining the ***sparsity pattern*** of $H^{-1} A H$, that is, which of the 16 sub-blocks are *necessarily* zero (although some of the non-"zero" blocks may turn out to be zero-dimensional or even zero matrices, depending on how the basis vectors are chosen in (1.--4.) above.

First, note that $\mathcal{C}\cap\mathcal{N} = \mathcal{R}(V_{c\bar{o}})$ is $A$-invariant, so there exists $A_{c\bar{o}}$ such that $A V_{c\bar{o}} = V_{c\bar{o}} A_{c\bar{o}}$.

Second, noting that $\mathcal{C} = \mathcal{R}(V_c)$ is $A$-invariant where $V_c = \left[\begin{array}{cc} V_{co} & V_{c\bar{o}} \end{array}\right]$, there exists $A_c$ such that $A V_c = V_c A_c$.  But in light of the previous observation that $A V_{c\bar{o}} = V_{c\bar{o}} A_{c\bar{o}}$, we conclude

$$ A_c = \left[\begin{array}{cc} A_{co} & 0 \\ A_{c\times} & A_{c\bar{o}} \end{array}\right]. $$

Third, noting that $\mathcal{N} = \mathcal{R}(V_{\bar{o}})$ is $A$-invariant where $V_{\bar{o}} = \left[\begin{array}{cc} V_{c\bar{o}} & V_{\bar{c}\bar{o}} \end{array}\right]$, there exists $A_\bar{o}$ such that $A V_\bar{o} = V_\bar{o} A_\bar{o}$.  But in light of the original observation that $A V_{c\bar{o}} = V_{c\bar{o}} A_{c\bar{o}}$, we conclude

$$ A_\bar{o} = \left[\begin{array}{cc} A_{\bar{c}o} & 0 \\ A_{\bar{c}\times} & A_{\bar{c}\bar{o}} \end{array}\right]. $$

Finally, since $V_{\bar{c}o}$ were chosen arbitrarily, there are no zeros automatically populated in the third column of $H^{-1} A H$.

To obtain the given structures for $B$ and $C$, we note that: $\mathcal{R}(B)\subset\mathcal{C} = \mathcal{R}(V_c)$, so there exists $B_c$ such that $B = V_c B_c$; $\mathcal{R}(C^\top) \subset \mathcal{N}^\perp$, so there exist $C_{co}$, $C_{\bar{c}o}$ such that $C H = C_{co} V_{co} + C_{\bar{c}o} V_{\bar{c}o}$.
